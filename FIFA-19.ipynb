```python
#To find best Goal Keeper
#Taking data for GK only 
best_team=[]
temp_all_GK=fifa.loc[(fifa["Position"]=="GK")] 
max_val=temp_all_GK["Overall"].max()
max_Potential=temp_all_GK["Potential"].max()
best_GK=fifa[(fifa["Overall"]==max_val) & (fifa["Potential"]==max_Potential)]
GK_Name=best_GK.iloc[0].loc['Name']
best_team.append(GK_Name)
print(best_team)
```

```python
#To find best Left Back
#Taking data for Left Back only
temp_all_LB=fifa.loc[(fifa["Position"]=="LB")]
max_val=temp_all_LB["Overall"].max()
max_Potential=temp_all_LB["Potential"].max()
best_LB=fifa[(fifa["Overall"]==max_val) & (fifa["Potential"]==max_Potential)]
Best_LB_Name=best_LB.iloc[0].loc['Name']
best_team.append(Best_LB_Name)
print(best_team)
```

```python
#To find best Right Back
#Taking data for Right Back only
temp_all_RB=fifa.loc[(fifa["Position"]=="RB")]
max_val=temp_all_RB["Overall"].max()
max_Potential=temp_all_RB["Potential"].max()
best_RB=fifa[(fifa["Overall"]==max_val) & (fifa["Potential"]==max_Potential)]
Best_RB_Name=best_RB.iloc[0].loc['Name']
best_team.append(Best_RB_Name)
print(best_team)
```

```python
#To find best Left Midfield
#Taking data for best left midfield player
temp_all_LM=fifa.loc[(fifa["Position"]=="LM")]
max_val=temp_all_LM["Overall"].max()
max_Potential=temp_all_LM["Potential"].max()
best_LM=fifa[(fifa["Overall"]==max_val) & (fifa["Potential"]==max_Potential)]
Best_LM_Name=best_LM.iloc[0].loc['Name']
best_team.append(Best_LM_Name)
print(best_team)
```

```python
#To find best Right Midfield
#Taking data for best left midfield player
temp_all_RM=fifa.loc[(fifa["Position"]=="RM")]
max_val=temp_all_RM["Overall"].max()
max_Potential=temp_all_RM["Potential"].max()
best_RM=fifa[(fifa["Overall"]==max_val) & (fifa["Potential"]==max_Potential)]
Best_RM_Name=best_RM.iloc[0].loc['Name']
best_team.append(Best_RM_Name)
print(best_team)
```

```python
#To find 2 best Center Back
#Taking data for 2 best left midfield player
temp_all_CB=fifa.loc[(fifa["Position"]=="CB")]
max_val=temp_all_CB["Overall"].max()
max_Potential=temp_all_CB["Potential"].max()
temp_best_CB=fifa[(fifa["Overall"]==max_val) | (fifa["Potential"]==max_Potential)]
Best_CB_Name=temp_best_CB.iloc[0:2]
best_team.append(Best_CB_Name.iloc[0].loc['Name'])
best_team.append(Best_CB_Name.iloc[1].loc['Name'])
print(best_team)
```

```python
#To find 2 best Center Midfield
#Taking data for 2 best center midfield player
temp_all_CM=fifa.loc[(fifa["Position"]=="CM")]
max_val=temp_all_CM["Overall"].max()
max_Potential=temp_all_CM["Potential"].max()
temp_best_CM=fifa[(fifa["Overall"]==max_val) | (fifa["Potential"]==max_Potential)]
Best_CM_Name=temp_best_CM.iloc[0:2]
best_team.append(Best_CM_Name.iloc[0].loc['Name'])
best_team.append(Best_CM_Name.iloc[1].loc['Name'])
print(best_team)
```

```python
#To find 2 best Center Forward
#Taking data for 2 best center Forward player
temp_all_CF=fifa.loc[(fifa["Position"]=="CF")]
max_val=temp_all_CF["Overall"].max()
max_Potential=temp_all_CF["Potential"].max()
temp_best_CF=fifa[(fifa["Overall"]==max_val) & (fifa["Potential"]==max_Potential)]
Best_CF_Name=temp_best_CF.iloc[0:2][['ID','Name']]
best_team.append(Best_CF_Name.iloc[0].loc['Name'])
best_team.append(Best_CF_Name.iloc[1].loc['Name'])
print(best_team)
```

```python
#Analyse which Club or National Team has the best-rated players
best_club=fifa.groupby('Club').sum().sort_values(["Overall"],ascending=False)
print("Best Club with highest highest overall ratings of players is ",best_club.index[0])
```

```python
#Analyse the team with the best dribbling speed

teamwise_drib_speed=fifa.groupby('Nationality')['Dribbling'].sum()
best_team_dribbling_speed=pd.DataFrame(data=teamwise_drib_speed)#,index=['teams'],columns=['Dribbling Speed'])
best_team=best_team_dribbling_speed.sort_values(["Dribbling"],ascending=False)
print(best_team.iloc[0])
```

```python
#Assess the strength of a team at a particular position
teamwise_pos_data=fifa.groupby('Nationality')['Positioning'].sum()
best_team_Pos=pd.DataFrame(data=teamwise_pos_data)#,index=['teams'],columns=['Dribbling Speed'])
best_team_Positioning=best_team_Pos.sort_values(["Positioning"],ascending=False)
print(best_team_Positioning.iloc[0])
```
